# Apartment Prices

```python
import pandas as pd
```

### Read in data

```python
df = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/Booli_sold.csv")
df
```

### Select

```python
relevant_df1 = df[["booliId", "soldPrice", "livingArea"]]
relevant_df1
```

## 1.Calculate the price per square meter (ppsqm) for each apartment sold. 

```python
## That is, add a new variable to your table containing the data representing the ppsqm
relevant_df1['ppsqm'] = relevant_df1['soldPrice'] / relevant_df1['livingArea']
relevant_df1
```

## 2. Rank the top 5 most expensive appartments w.r.t ppsqm. 
## &nbsp;&nbsp;&nbsp;&nbsp;Present this in a table.

```python
sorted_df = relevant_df1.sort_values("ppsqm", ascending=False)
sorted_df.head()
```

## 3. What is the avereage ppsqm in Ekhagen?

```python
Average_ppsqm = relevant_df1['ppsqm'].mean()
print("The avereage ppsqm in Ekhagen is", Average_ppsqm)
```

## 4. The aspect of the data that interests me 

```python
relevant_df2 = df[["booliId", "soldPrice", "listPrice"]]
relevant_df2
```

```python
row_count_relevant_df2 = len(relevant_df2)
row_count_relevant_df2
```

```python
Part_relevant_df2 = relevant_df2[relevant_df2['soldPrice'] < relevant_df2['listPrice']]
Part_relevant_df2
```

```python
row_count_Part_relevant_df2 = len(Part_relevant_df2)
row_count_Part_relevant_df2
```
