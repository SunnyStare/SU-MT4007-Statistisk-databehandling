```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
url = "https://gymnasieantagningen.storsthlm.se/slutantagning/"
driver.get(url)

try:
    # 强制移除干扰元素
    driver.execute_script("""
        document.querySelectorAll('#cookieDetails, #cookiebanner, #top-nav, img[alt="Cookie logo"], label, p').forEach(function(el) {
            el.remove();
        });
    """)

    # 确认元素是否移除
    print("Checking if key interference elements are removed:")
    for selector in ['#cookieDetails', '#cookiebanner', '#top-nav', "img[alt='Cookie logo']", "label", "p"]:
        element = driver.execute_script(f"return document.querySelector('{selector}')")
        print(f"{selector}: {'Exists' if element else 'Removed'}")

    # 点击年份选择菜单
    span_selection = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "span.selection"))
    )
    span_selection.click()

    # 选择年份 2020
    year_option = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//li[contains(@class, 'select2-results__option') and text()='2020']"))
    )
    year_option.click()

    print("Year 2020 selected successfully!")
except Exception as e:
    print(f"Exception occurred: {e}")
finally:
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 替换 'chromedriver_path' 为你的 ChromeDriver 的实际路径
service = Service('chromedriver_path')
driver = webdriver.Chrome(service=service)

try:
    # 打开目标网址
    driver.get('https://gymnasieantagningen.storsthlm.se/slutantagning/')

    # 这里继续你的逻辑，例如移除干扰元素、选择年份等
    driver.execute_script("""
        const elementsToRemove = document.querySelectorAll('#cookieDetails, #cookiebanner, #top-nav, img[alt="Cookie logo"], label, p');
        elementsToRemove.forEach(el => el.remove());
    """)
    
    # 点击年份下拉菜单
    span_selection = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "span.selection"))
    )
    span_selection.click()

    # 从动态选项中选择 2020
    year_option = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//li[contains(@class, 'select2-results__option') and text()='2020']"))
    )
    year_option.click()

    print("Year 2020 selected successfully!")
except Exception as e:
    print(f"Exception occurred: {e}")
finally:
    driver.quit()
```
