```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# IRIS data

```python
# Read in data
df = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/IRIS.csv")
df.info()
```

## Is there a relationship between sepal dimensions and petal dimensions?

```python
# Two graphs in one row
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

background_color = '#f0f0f0'

# Numerical vs. Numerical - sepal_length vs. petal_length                         
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, palette='Set1', ax=axes[0]) 
axes[0].set_title('The relationship between Sepal Length and Petal Length') 
axes[0].set_xlabel('Sepal Length') 
axes[0].set_ylabel('Petal Length')
axes[0].set_facecolor(background_color)
axes[0].grid(True, color='white', alpha=0.6)

# Numerical vs. Numerical - sepal_length vs. petal_length                         
sns.scatterplot(x='sepal_width', y='petal_width', hue='species', data=df, palette='Set1', ax=axes[1]) 
axes[1].set_title('The relationship between Sepal Width and Petal Width') 
axes[1].set_xlabel('Sepal Width') 
axes[1].set_ylabel('Petal Width')
axes[1].set_facecolor(background_color)
axes[1].grid(True, color='white', alpha=0.6)

# print
plt.tight_layout() 
plt.show()
```

### &nbsp;&nbsp;&nbsp;&nbsp;There are visually noticeable correlations 
### between sepal dimensions and petal dimensions 
### for iris-versicolor and iris-virginica, 
### but not for iris-setosa.

## How are the sepal and petal dimensions distributed? 

```python
# Create figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
background_color = '#f0f0f0'

# Plot the first subplot (Species vs. Sepal Width)
sns.boxplot(x='species', y='sepal_width', data=df, hue='species', palette='Set1', ax=axes[0, 0])
axes[0, 0].set_xlabel('Species')
axes[0, 0].set_ylabel('Sepal Width')
axes[0, 0].set_facecolor(background_color)
axes[0, 0].grid(True, color='white', alpha=0.6)

# Plot the second subplot (Species vs. Sepal length)
sns.boxplot(x='species', y='sepal_length', data=df, hue='species', palette='Set1', ax=axes[0, 1])
axes[0, 1].set_xlabel('Species')
axes[0, 1].set_ylabel('Sepal Length')
axes[0, 1].set_facecolor(background_color)
axes[0, 1].grid(True, color='white', alpha=0.6)

# Plot the third subplot (Species vs. Petal Width)
sns.boxplot(x='species', y='petal_width', data=df, hue='species', palette='Set1', ax=axes[1, 0])
axes[1, 0].set_xlabel('Species')
axes[1, 0].set_ylabel('Petal Width')
axes[1, 0].set_facecolor(background_color)
axes[1, 0].grid(True, color='white', alpha=0.6)

# Plot the fourth subplot (Species vs. Petal length)
sns.boxplot(x='species', y='petal_length', data=df, hue='species', palette='Set1', ax=axes[1, 1])
axes[1, 1].set_xlabel('Species')
axes[1, 1].set_ylabel('Petal Length')
axes[1, 1].set_facecolor(background_color)
axes[1, 1].grid(True, color='white', alpha=0.6)

# Adjust subplot layout
plt.tight_layout()
plt.show()
```

### &nbsp;&nbsp;&nbsp;&nbsp; The petals of Iris-setosa are the smallest among the three species, 
### and their size and shape are particularly consistent.

## How the different variables are related to each other, 
## &nbsp;&nbsp;&nbsp;&nbsp; the pair-plot of the three types of iris.

```python
selected_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
palette = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}
sns.pairplot(df[selected_features], hue='species', palette=palette)
plt.show()
```

### &nbsp;&nbsp;&nbsp;&nbsp; The ratio between the length and width of the sepals of Iris-setosa is relatively fixed, meaning that its shape is stable.The ratio between the length and width of the sepals in the other two species varies within a certain size range, meaning that their shape is less stable.
### &nbsp;&nbsp;&nbsp;&nbsp; The length and width of the petal of Iris-setosa are relatively fixed, meaning that both its shape and its size are stable. The length and width of the sepals in the other two species varies, but the ratio between them is relatively fixed. That means, the size of the other two species are relatively less stable, Iris-virginica are often bigger than Iris-versicolor, but the shape of petal of these two species are the same.

```python
# Read in data
df = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/artportalen.csv")
df.info()
df.head(50)
```

## What are the most prevelant species?

```python
df["Antal"].unique()
```

```python
# Select the relevant parts of the data and fix the data type
relevant_df2_1 = df[["Artnamn", "Antal"]].copy()
relevant_df2_1.loc[relevant_df2_1["Antal"] == "noterad"] = 1 
relevant_df2_1.loc[relevant_df2_1["Antal"] == "Ej terfunnen"] = 0
relevant_df2_1["Antal"] = relevant_df2_1["Antal"].astype('int64')
relevant_df2_1["Antal"].unique()
```

```python
# Sum and Sort
grouped_df2_1 = relevant_df2_1.groupby ("Artnamn").sum()
sorted_df2_1 = grouped_df2_1.sort_values("Antal", ascending=False)
sorted_df2_1.head(3)
```

## What is the monthly distribution of the top 3 most prevelant species?

```python
df["Slutdatum"].unique()
```

```python
# Select the relevant parts of the data 
relevant_df2_2 = df[["Artnamn", "Antal","Slutdatum"]].copy()
# Fix the data type
relevant_df2_2["Antal"].loc[relevant_df2_2["Antal"] == "noterad"] = 1
relevant_df2_2["Antal"].loc[relevant_df2_2["Antal"] == "Ej terfunnen"] = 0
relevant_df2_2["Antal"] = relevant_df2_2["Antal"].astype('int64')
# Seperate data in month
relevant_df2_2['Slutdatum'] = pd.to_datetime(relevant_df2_2['Slutdatum'])
relevant_df2_2['Month'] = relevant_df2_2['Slutdatum'].dt.month
relevant_df2_2
```

```python
# Gronsiska
Gronsiska = relevant_df2_2.loc[(relevant_df2_2["Artnamn"]=="Grnsiska")]
sns.histplot(Gronsiska['Month'])
plt.xlabel('Month') 
plt.ylabel('Antal') 
plt.title('Monthly Distribution of Grnsiska') 
plt.show()
```

```python
# Sothna
Sothona = relevant_df2_2.loc[(relevant_df2_2["Artnamn"]=="Sothna")]
sns.histplot(Sothona['Month'])
plt.xlabel('Month') 
plt.ylabel('Antal') 
plt.title('Monthly Distribution of Sothna') 
plt.show()
```

```python
# Grsand
Grasand = relevant_df2_2.loc[(relevant_df2_2["Artnamn"]=="Grsand")]
sns.histplot(Grasand['Month'])
plt.xlabel('Month') 
plt.ylabel('Antal') 
plt.title('Monthly Distribution of Grsand') 
plt.show()
```

## What are the rarest species?

```python
# Sum and Sort
grouped_df2_3 = relevant_df2_1.groupby ("Artnamn").sum()
Rarest = grouped_df2_3.loc[(grouped_df2_3["Antal"]==1)]
Rarest
```

## Where are the rarest species? Perhaps we could allocate more resources to these areas to protect these birds.

```python
# Select the relevant parts of the data and fix the data type
relevant_df2_4 = df[["Artnamn", "Antal","Kommun"]].copy()
relevant_df2_4["Antal"].loc[relevant_df2_4["Antal"] == "noterad"] = 1
relevant_df2_4["Antal"].loc[relevant_df2_4["Antal"] == "Ej terfunnen"] = 0
relevant_df2_4["Antal"] = relevant_df2_4["Antal"].astype('int64')
relevant_df2_4["Kommun"].unique()
```

```python
grouped_df2_4 = relevant_df2_4.groupby ("Artnamn").sum()
Rarest = grouped_df2_4.loc[(grouped_df2_4["Antal"]==1)]
Rarest
```

```python
Kommun_Rarest = Rarest.groupby ("Kommun").sum()
# Calculate the percentage 
total_counts = Kommun_Rarest["Antal"].sum() 
Kommun_Rarest["Percentage"] = (Kommun_Rarest["Antal"] / total_counts) * 100

Kommun_Rarest.sort_values("Antal", ascending=False)
```

### Statistics show that Stockholm has the most rarest bird species.

## But, maybe, some species of birds are not rare, but are just less common in Stockholm?

```python
# Select the relevant parts of the data and fix the data type
relevant_df2_5 = df[["Artnamn", "Antal","Kommun","Rdlistade"]].copy()
relevant_df2_5["Antal"].loc[relevant_df2_5["Antal"] == "noterad"] = 1
relevant_df2_5["Antal"].loc[relevant_df2_5["Antal"] == "Ej terfunnen"] = 0
relevant_df2_5["Antal"] = relevant_df2_5["Antal"].astype('int64')
relevant_df2_5["Rdlistade"].unique()
```

```python
# Select all the threatened species, 
# i.e. those endangered(EN) and vulnerable(VU) in the column "Rdlistade".
Rare = relevant_df2_5.loc[(relevant_df2_5["Rdlistade"]=='EN') | (relevant_df2_5["Rdlistade"]=='VU')]
Rare
```

```python
Antal_Kommun_Rare = Rare[["Antal","Kommun"]].copy()
Kommun_Antal_Rare = Antal_Kommun_Rare.groupby ("Kommun").sum()
# Calculate the percentage column 
total_count = Kommun_Antal_Rare["Antal"].sum() 
Kommun_Antal_Rare["Percentage"] = (Kommun_Antal_Rare["Antal"] / total_count) * 100

Kommun_Antal_Rare.sort_values("Antal", ascending=False)
```

### Statistics show again that Stockholm has the most rare bird species.

## One more time, but, maybe, it is only because there are more observations in Stockholm?

```python
# Count the number of rows for each unique value
Observation_Kommun = df["Kommun"].value_counts().reset_index() 
# Rename columns 
Observation_Kommun.columns = ["Kommun", "Counts"] 
# Calculate the percentage 
total_counts = Observation_Kommun["Counts"].sum() 
Observation_Kommun["Percentage"] = (Observation_Kommun["Counts"] / total_counts) * 100
# Print the resulting dataframe 
Observation_Kommun
```

### &nbsp;&nbsp;&nbsp;&nbsp; Oh! No! It appears that the majority of rare bird species recorded in the data are from Stockholm, primarily because most of the observation records come from Stockholm. Therefore, this data source is not ideal for studying where to invest resources to protect rare birds. Next time, it would be better to mobilize observers outside of Stockholm to be more diligent in their observations.
